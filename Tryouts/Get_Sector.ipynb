```python
def get_sector(ticker_name):
    type_of_sector = []
    url = "https://finance.yahoo.com/quote/"+ticker_name+"/profile?p="+ticker_name
    res = requests.get(url)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'html.parser')
    for hit in soup.findAll(attrs={'class' : 'Fw(600)'}):
        type_of_sector.append(hit.text.strip())
    if len(type_of_sector)>0:
        return type_of_sector[0],';',
    else:
        return "Not Found"
```

```python
def get_industry(ticker_name):
    type_of_industry = []
    url = "https://finance.yahoo.com/quote/"+ticker_name+"/profile?p="+ticker_name
    res = requests.get(url)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'html.parser')
    for hit in soup.findAll(attrs={'class' : 'Fw(600)'}):
        type_of_industry.append(hit.text.strip())
    if len(type_of_industry)>0:
        return type_of_industry[1]
    else:
        return "Not Found"
```
